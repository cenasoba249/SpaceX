```python
# SpaceX Falcon 9 First Stage Landing Prediction - 完全版コード

# 必要なライブラリのインストール（PyLite環境用）
import piplite
await piplite.install(['numpy'])
await piplite.install(['pandas'])
await piplite.install(['seaborn'])

# ライブラリのインポート
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from js import fetch
import io

# データの読み込み
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
resp = await fetch(URL)
dataset_part_2_csv = io.BytesIO((await resp.arrayBuffer()).to_py())
df = pd.read_csv(dataset_part_2_csv)

# データの確認
print("データの形状:", df.shape)
print("データの最初の5行:")
df.head(5)

# 特徴量データフレームの作成
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]

print("特徴量データフレームの形状:", features.shape)
print("特徴量データフレームの最初の5行:")
features.head()

# =============================================================================
# TASK 7: カテゴリカル列のダミー変数作成
# =============================================================================

print("\n=== TASK 7: One-Hot Encoding の実行 ===")

# カテゴリカル列を指定
categorical_columns = ['Orbit', 'LaunchSite', 'LandingPad', 'Serial']

# One-Hot Encodingを実行
features_one_hot = pd.get_dummies(features, columns=categorical_columns, dtype=int)

print("One-Hot Encoding後の形状:", features_one_hot.shape)
print("One-Hot Encoding後の列数:", len(features_one_hot.columns))
print("One-Hot Encoding後の最初の5行:")
features_one_hot.head()

# カテゴリカル列ごとの詳細分析
print("\n=== カテゴリカル列の詳細分析 ===")
for col in categorical_columns:
    unique_values = features[col].nunique()
    print(f"{col}: {unique_values} ユニーク値")
    print(f"  値: {features[col].unique()}")

# =============================================================================
# TASK 8: 全ての数値列をfloat64にキャスト
# =============================================================================

print("\n=== TASK 8: データ型をfloat64に変換 ===")

# データ型の確認（変換前）
print("変換前のデータ型:")
print(features_one_hot.dtypes.value_counts())

# 全列をfloat64に変換
features_one_hot = features_one_hot.astype('float64')

print("\n変換後のデータ型:")
print(features_one_hot.dtypes.value_counts())

# =============================================================================
# 可視化の例（クイズ回答用）
# =============================================================================

print("\n=== 可視化の実行 ===")

# PayloadMass vs FlightNumber のプロット
plt.figure(figsize=(12, 6))
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect=5)
plt.xlabel("Flight Number", fontsize=20)
plt.ylabel("Pay load Mass (kg)", fontsize=20)
plt.title("PayloadMass vs FlightNumber by Class")
plt.show()

# LaunchSite vs FlightNumber のスキャッタープロット（Question 3用）
plt.figure(figsize=(12, 6))
sns.catplot(y="LaunchSite", x="FlightNumber", hue="Class", data=df, aspect=1, kind='scatter')
plt.ylabel("Launch Site", fontsize=15)
plt.xlabel("Flight Number", fontsize=15)
plt.title("LaunchSite vs FlightNumber Scatterplot")
plt.show()

# バーチャートの例（Question 1用）
plt.figure(figsize=(10, 6))
launch_site_counts = df['LaunchSite'].value_counts()
plt.bar(launch_site_counts.index, launch_site_counts.values)
plt.xlabel("Launch Site")
plt.ylabel("Count")
plt.title("Launch Site Distribution (Bar Chart for Categorical Data)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# =============================================================================
# データの出力
# =============================================================================

# CSVファイルへの出力
features_one_hot.to_csv('dataset_part_3.csv', index=False)

print("\n=== 最終結果 ===")
print(f"最終的なデータフレームの形状: {features_one_hot.shape}")
print(f"列数: {len(features_one_hot.columns)}")
print(f"行数: {len(features_one_hot)}")
print("dataset_part_3.csvに保存されました。")

# 列名の表示
print("\n全列名:")
for i, col in enumerate(features_one_hot.columns):
    print(f"{i+1:2d}. {col}")

print("\n処理完了！")
```
